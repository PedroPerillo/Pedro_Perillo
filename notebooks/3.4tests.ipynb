```python
%sql
-- Steps vs. No-Steps
SELECT step_label, COUNT(*)
FROM silver.labeled_step_test
GROUP BY step_label;
```

```python
%sql
-- Invalid or missing labels
SELECT *
FROM silver.labeled_step_test
WHERE step_label NOT IN ('step', 'no_step')
  OR step_label IS NULL


LIMIT 50;
```

```python
%sql
-- Source label counts
SELECT source, COUNT(*)
FROM silver.labeled_step_test
GROUP BY source;
```

```python
%sql
-- Invalid source labels
SELECT *
FROM silver.labeled_step_test
WHERE source NOT IN ('device', 'step')
   OR source IS NULL
LIMIT 50;
```
